```python
#Data Pre-Processing
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
correlation = parkinsons_data.corr()
```

```python
# constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot = True, annot_kws={'size':8}, cmap =  'Blues')
```

```python
df = parkinsons_data.dropna()
df.isna().sum()
```

```python
sns.pairplot(df,diag_kind="kde")
plt.show()
```

```python
#Splitting the data to training data & Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

SVM 

```python
#Support Vector Machine Model
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy_svc = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy_svc)
```

```python
# accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy_svc = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy_svc)
```

```python
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)


if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("The Person has Parkinsons")
```
